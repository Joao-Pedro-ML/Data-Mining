#### 1- Faça uma busca por 100 vídeos com a palavra chave 'eleição'. Faça a procura por conteúdos no Brasil no período do segundo turno da última eleição presidencial. Identifique o vídeo com mais comentários e exiba as informações desse vídeo.

```python
from pyyoutube import Api

api = Api(api_key="")

r = api.search_by_keywords(
    q="eleição", 
    search_type=["channel","video", "playlist"], 
    count=100,
    limit=100, 
    published_after="2022-10-02T00:00:00Z", 
    published_before="2022-10-30T00:00:00Z"
)

max_comments = -1
video_max_comments = None

for item in r.items:
    video_id = item.id.videoId
    video_details = api.get_video_by_id(video_id=video_id).items[0]

    comment_count = video_details.statistics.commentCount

    if comment_count is None:
        comment_count = 0
    else:
        comment_count = int(comment_count)

    if comment_count > max_comments:
        max_comments = comment_count
        video_with_max_comments = video_details

print(f"Vídeo com mais comentários: {video_with_max_comments.snippet.title}")
print(f"Número de comentários: {max_comments}")
print(f"Link: https://www.youtube.com/watch?v={video_with_max_comments.id}")
video_with_max_comments.to_dict()
```

#### 3- Crie uma aplicação que usa o Selenium para abrir o site https://www.uol.com.br/ e rolar a página para baixo até o fim do conteúdo.

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time


chrome_driver_path = 'chromedriver-win64/chromedriver.exe'
service = Service(chrome_driver_path)

driver = webdriver.Chrome(service=service)
driver.get("https://www.uol.com.br/") 

time.sleep(5)


last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    time.sleep(2)

    new_height = driver.execute_script("return document.body.scrollHeight")

    if new_height == last_height:
        break

    last_height = new_height

driver.quit()
```
